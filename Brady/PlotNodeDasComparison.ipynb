```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import matplotlib.ticker as ticker
sns.set_style('ticks')
%matplotlib inline
```

```python
coloro = "#FF6C0C"
colorb = "#00A1DF"
colorgr = "#C8C8C8"
colorp = "#F54D80"
```

```python
nodes_only = np.loadtxt("nodes_only.txt")
w_das = np.loadtxt("das_w_nodes.txt")
```

```python
percents = [20,40,60,80,100]
das_percents = [0,20,40,60,80,100]
```

```python
fig, ax = plt.subplots(figsize=(8,6))

for i, path in enumerate(nodes_only.T):
    if i == 0:
        ax.plot(percents, path, lw=0.5, marker='o', c=colorb, label="Nodes Only")
    else:
        ax.plot(percents, path, lw=0.5, marker='o', c=colorb)
        
for i, path in enumerate(w_das.T):
    if i == 0:
        ax.plot(das_percents, path, lw=0.5, marker='s', c=coloro, label="Nodes and DAS")
    else:
        ax.plot(das_percents, path, lw=0.5, marker='s', c=coloro)
        
ax.legend(loc="upper right")
ax.set_xticks(das_percents)
ax.set_xlim(0, 101)
ax.set_ylim(0, 1.3)
ax.set_ylabel("Node Validation MSE (normalized)", fontsize=14)
ax.set_xlabel("Training Nodes Utilized (%)", fontsize=14)
plt.savefig("NodeDasCVComparison.pdf")
```
